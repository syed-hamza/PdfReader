```python
import tabula
from tabulate import tabulate
import pandas as pd

try:
    # Read table from PDF file
    dfs = tabula.read_pdf("./FAPDF/1706.03762v7.pdf", pages="all", multiple_tables=True, encoding='ISO-8859-1')  # specify encoding if known
    
    # Check if the result is a list of DataFrames
    if isinstance(dfs, list) and all(isinstance(df, pd.DataFrame) for df in dfs):
        for df in dfs:
            print(tabulate(df, headers='keys', tablefmt='psql'))
    else:
        print("No tables found in the PDF.")
except UnicodeDecodeError as e:
    print(f"UnicodeDecodeError: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```
